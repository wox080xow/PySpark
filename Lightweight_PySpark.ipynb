```python
from pyspark.sql import SparkSession
import pyspark.sql.functions as fn
```

```python
# local mode
spark = SparkSession \
        .builder \
        .appName("LocalMode") \
        .getOrCreate()
```

```python
# yarn mode
spark = SparkSession \
        .builder \
        .master("yarn") \
        .config('spark.executor.instances','100') \
        .config('spark.executor.memory','4G') \
        .config('spark.executor.cores','1') \
        .appName("YARN") \
        .getOrCreate()
```

```python
# standalone mode
spark = SparkSession \
        .builder \
        .master("spark://bdse120.example.org:7077") \
        .config('spark.cores.max','1') \
        .config('spark.executor.memory','1G') \
        .appName("Standalone") \
        .getOrCreate()
```

```python
%%time
x = 100000
count=0
for n in range(1, x): 
    c = 2 
    while c < n: 
        if n % c == 0: 
            break 
        c += 1 
    else: 
        count = count+1 
print('There are', count, 'prime numbers between 1 to', x) 
```
